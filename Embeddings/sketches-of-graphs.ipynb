**IMPORTING LIBRARY**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# SKETCH of GRAPH

```python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G'),('A','P'),('A','L'),('L','M'),('L','N')])

plt.axis('off')
nx.draw_networkx(G,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=800,
                 cmap='#36454F',  
                 font_size=15,
                 font_color='#FF9933'
                 )
# cmap - #36454F -> Charcoal Gray .
# font_color - #FF9933 -> saffron .
```

**DIRECTED GRAPH**

```python
DG = nx.DiGraph()
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G'),('A','P'),('A','L'),('L','M'),('L','N')])

plt.axis('off')
nx.draw_networkx(DG,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=800,
                 cmap='#36454F',
                 font_size=15,
                 font_color='#FF9933'
                 )
```

**WEIGHTED GRAPH**

```python
WG = nx.Graph()
WG.add_edges_from([('A', 'B', {"weight": 15}), ('A', 'C', {"weight": 80}), ('B', 'D', {"weight": 20}), ('B', 'E', {"weight": 90}), ('C', 'F', {"weight": 50}), ('C', 'G', {"weight": 60}),('A','P',{"weight": 40}),('A','L',{"weight": 75}),('L','M',{"weight": 45}),('L','N',{"weight": 25})])
labels = nx.get_edge_attributes(WG, "weight")

plt.axis('off')
nx.draw_networkx(WG,
                 pos=nx.spring_layout(G, seed=0),
                 node_size=800,
                 cmap='#36454F',
                 font_size=15,
                 font_color='#FF9933'
                 )
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(G, seed=0), edge_labels=labels)
```

```python
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 1), (1, 4)])
print(f"Is graph 1 connected? {nx.is_connected(G1)}")

G2 = nx.Graph()
G2.add_edges_from([(1, 2), (2, 3), (3, 1), (4, 5),(5,6),(4,6)])
print(f"Is graph 2 connected? {nx.is_connected(G2)}")
```
